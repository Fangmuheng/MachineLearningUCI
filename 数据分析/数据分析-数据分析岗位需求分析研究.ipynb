```python
#缺失值检测函数
def null_value_check(data):
    '''输入：pandas数组'''
    '''输出：每一列的缺失值情况'''
    for i in data.columns:
        print(i,'列缺失值个数为：',data[i].isnull().sum())
```

```python
#检查一下有没有缺失值
null_value_check(position)
```

对于缺失值的处理方式通常有：<br>
1.删除这些变量是缺失值(unknown)的行；<br>
2.如果预计该变量对于学习模型效果影响不大，可以对unknown值赋众数，这里认为变量都对学习模型有较大影响，不采取此法；<br>
3.可以使用数据完整的行作为训练集，以此来预测缺失值。<br>

<br><br>
但我们只是对数据进行分析，并不对数据进行建模，且缺失值不多，对其他行没有影响。在这里我们直接删掉有缺失值的行即可。

```python
#删除缺失值
#axis=0: 删除包含缺失值的行
#how=‘any’ :只要有缺失值出现，就删除该行
#inplace=True在原数据上操作
position.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
```

<B>（3）薪资水平统计-该岗位一般处于什么薪资水平？</B>

首先我们对表格里的薪资水平格式进行修改。因为表格里以范围的形式显示，这里我们取范围中值进行分析。

```python
def salary_count(x):
    x_min,x_max = x.split('-')  #获取-两边的最大最小值
    x_min = int(x_min.replace('k','000'))
    x_max = int(x_max.replace('k','000')) #截取字符串，把k改成000，将字符串转化为数字
    return np.mean([x_min,x_max])
```

```python
#选取北上广深四个城市进行岗位平均工资分析

city = ['北京','上海','广州','深圳']

salary1 = position[position['city']=='北京']
salary2 = position[position['city']=='上海']
salary3 = position[position['city']=='广州']
salary4 = position[position['city']=='深圳']
```

```python
salary1 = np.array(salary1['salary'].apply(salary_count)).mean()
salary2 = np.array(salary2['salary'].apply(salary_count)).mean()
salary3 = np.array(salary3['salary'].apply(salary_count)).mean()
salary4 = np.array(salary4['salary'].apply(salary_count)).mean()
```

```python
salary = [salary1,salary2,salary3,salary4]
salary
```

```python
bar=pyec.Bar()
bar.add_xaxis(city)
bar.add_yaxis('平均工资',salary)
bar.render_notebook()
```

从图表中我们可以看出，北京行业平均工资最高，达23016元，广州在 四个城市里最低，只有15523元。
